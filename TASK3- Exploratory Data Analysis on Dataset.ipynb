```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#Loading Or Reading The CSV File
Terrorism_Dataframe = pd.read_csv("C:\globalterrorismdb_0718dist.csv",encoding='latin1',low_memory=False)
```

```python
#Renaming all Columns for understanding

Terrorism_Dataframe.rename(columns={'iyear':'Year','imonth':'Month','iday':"Day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'Summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'Kill',
     'nwound':'Wound'},inplace=True)
```

```python
# Shows the Statistical details of the column which has numerical values.
Terrorism_Dataframe.describe() 
```

```python
# Gives the count of categorical column with respect to each category.
pd.DataFrame(Terrorism_Dataframe.City).value_counts()
```

```python
# Gives the count of categorical column with respect to each category.
pd.DataFrame(Terrorism_Dataframe.Country).value_counts() 
```

```python
#Dropping the unnecessary columns
Terrorism_Dataframe.drop(['motive','Summary','latitude','longitude'],axis=1,inplace=True)
```

```python
#It shows the shape of the data variable in tuple format i.e. no. of rows and column.
Terrorism_Dataframe.shape
```

```python
#Shows the Statistical details of the column which has numerical values.
Terrorism_Dataframe.describe()
```

```python
#Data Visualization
# Correlation Matrix
correlation = Terrorism_Dataframe.corr()
correlation
```

```python
#Visualize the Correlation matrix
sns.heatmap(correlation,annot=True)
```

```python
# Displays first 10 rows from value counts w.r.t country column
attack = Terrorism_Dataframe.Country.value_counts()[:10] 
attack
```

```python
year = Terrorism_Dataframe['Year'].unique()
years_count = Terrorism_Dataframe['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (16,8))
sns.barplot(x = year,y = years_count,palette = "Paired")
plt.xticks(rotation=90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Attacking Year',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=25)
plt.title('Attacks Analysis',fontsize=40)
plt.show()
```

```python
df = Terrorism_Dataframe[['Year','Kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,15))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.title("People Died Analysis",fontsize=25)
plt.ylabel("Number of killed people",fontsize=20)
plt.xlabel('Year',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
pd.crosstab(Terrorism_Dataframe.Year, Terrorism_Dataframe.Region).plot(kind='area',stacked=False,figsize=(10,5))
plt.title('Terrorist Activities By Region In Each Year',fontsize=15)
plt.ylabel('Number of Terrorist Attacks',fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(10,5))
sns.barplot(x=Terrorism_Dataframe['Country'].value_counts()[:10].index,y=Terrorism_Dataframe['Country'].value_counts()[:10].values,palette="Set2")
plt.title('Counteries Affected By Attacks',fontsize=10)
plt.xlabel('Countries',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
Terrorism_Dataframe[['Attacktype','Kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(10,5),color="Pink")
plt.xticks(rotation=30,fontsize=10)
plt.yticks(fontsize=10)
plt.title("Number of Killed vs Type of Attacks",fontsize=10)
plt.ylabel('Number of people killed',fontsize=10)
plt.xlabel('Type of Attacks',fontsize=10)
plt.show()
```

```python
Terrorism_Dataframe['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(10,5),color="Orange")
plt.xticks(rotation = 30,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("City",fontsize=10)
plt.ylabel("Number of attack",fontsize=10)
plt.title("Top 10 most Affected city",fontsize=10)
plt.show()
```

```python
Terrorism_Dataframe['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='Yellow',figsize=(10,5))
plt.title("Top 10 Terrorist group Attack",fontsize=10)
plt.xlabel("Name of Terrorist Group",fontsize=10)
plt.ylabel("Number of Attacks",fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
plt.subplots(figsize=(10,5))
sns.countplot(x=Terrorism_Dataframe["Targettype"],order=Terrorism_Dataframe['Targettype'].value_counts().index,palette="winter",edgecolor=sns.color_palette("winter"));
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("Type of Target",fontsize=10)
plt.ylabel("count",fontsize=10)
plt.title("Attack per Year",fontsize=10)
plt.show()
```
